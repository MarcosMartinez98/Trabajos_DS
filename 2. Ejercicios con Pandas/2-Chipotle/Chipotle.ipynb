### Step 1. Import the necessary libraries

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep = '\t')
df
```

### Step 3. Name of the max valued product

```python
def formatear(value):
    return float(value[1:])

df.loc[:,'item_price'] = df.loc[:,'item_price'].astype(str).apply(formatear)
df.loc[:,'item_price']
# si quiero que la nueva columna tengo tipo float he de crear una nueva
# df.loc[:,'item_price_float'] = df.loc[:,'item_price'].astype(str).apply(formatear)
# con bucle for siempre funciona
# precios_float = []
# for index, row in df.iterrows():
#     num_string = (row['item_price'].replace('$', ''))
#     num_float = float(num_string)
#     precios_float.append(num_float)
# df.loc[:, 'item_price_float'] = precios_float
```

```python
df.groupby('item_name')['item_price'].max().sort_values(ascending=False).head(1)
```

### Step 4. How many products cost more than $10.00?

```python
mask = df.loc[:,'item_price'] > 10
df.loc[mask,:'item_price']['item_name'].count()
#df.loc[mask,:].shape[0]
```

Extra: Cuántos pedidos se han hecho con un producto de más de 10$

```python
df.loc[mask,:'order_id'].nunique()
```

### Step 5. What is the price of each item? Get only item_name and unit_price

```python
# Crea una nueva columna llamada unit_price
series = df.groupby('item_name')['item_price'].mean()
df2 = series.to_frame()
df2 = df2.reset_index()

df3 = pd.merge(df,df2, how='inner', left_on = 'item_name', right_on = 'item_name')
df3.columns = ['order_id', 'quantity', 'item_name', 'choice_description','item_price', 'mean_item_price']
df3

```

### Step 6. Sort by the name of the item

```python
df3.sort_values('item_name').groupby('item_name')['mean_item_price'].max().to_frame().reset_index()
df3.sort_values('item_name').groupby('item_name')[['mean_item_price']].max().reset_index() # hace lo mismo que el to_frame()
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
# # # vamos primero a conseguir el producto mas caro
df3.sort_values('item_price', ascending=False)
# obtenemos que el producto mas caro es:  Chips and Fresh Tomato Salsa
mask = df3.loc[:,'item_name'] == 'Chips and Fresh Tomato Salsa'
df3.loc[mask,:].count()
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
# creamos una mascara para obtener el df filtrado
mask = df.loc[:,'item_name'] == 'Veggie Salad Bowl'
df.loc[mask,:].count().max()
#mask.sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
mask = (df.loc[:,'item_name'] == 'Canned Soda') & (df.loc[:,'quantity'] > 1)
df.loc[mask,:].count().max()
```
