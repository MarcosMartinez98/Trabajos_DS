```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|', index_col='user_id')
users 
```

```python
users.info()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean().to_frame()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.groupby('occupation')['gender']
```

```python
# esto me da el numero de personas por ocupacion, he de hallar el numero de hombres por ocupacion y dividirlo entre esto
users.groupby('occupation')['gender'].count().to_frame()
# para ello, vamos a hacer una mascara
mask = users.loc[:,'gender'] == 'M'
df_hombres = users.loc[mask,:]
df_hombres.groupby('occupation')['gender'].count().to_frame()

ratio = (df_hombres.groupby('occupation')['gender'].count().to_frame()).div(users.groupby('occupation')['gender'].count().to_frame())
ratio.reset_index()

```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
minimo = users.groupby('occupation')['age'].min().to_frame().reset_index()
maximo = users.groupby('occupation')['age'].max().to_frame().reset_index()

solucion = pd.merge(minimo,maximo, left_on='occupation', right_on='occupation', how='inner')
solucion.columns = ['occupation', 'age_min', 'age_max']
solucion
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
# vamos a calcular la edad media para hombres
mask = users.loc[:,'gender'] == 'M'
users_hombres = users.loc[mask,:]
df1 = users_hombres.groupby('occupation')['age'].mean().to_frame().reset_index()

# vamos ahora con las mujeres
mask = users.loc[:,'gender'] == 'F'
users_hombres = users.loc[mask,:]
df2 = users_hombres.groupby('occupation')['age'].mean().to_frame().reset_index()

df3 = pd.merge(df1,df2,left_on='occupation', right_on='occupation',how='inner')
df3.columns = ['occupation','male_age_mean','female_age_mean']
df3
```

### Step 8.  For each occupation present the percentage of women and men

```python
# vamos a calcular el numero total de personas por ocupacion
people_occ = users.groupby('occupation')['occupation'].count().to_frame()
# ahora no puedo hacer reset_index porque las columnas se llaman igual vamos a renombrarlas
people_occ.columns = ['num_people']
people_occ = people_occ.reset_index()
people_occ
# ahora vamos a hacer lo mismo para hombres 
mask = users.loc[:,'gender'] == 'M'
users_hombres = users.loc[mask,:]
hombres_occ = users_hombres.groupby('occupation')['occupation'].count().to_frame()
hombres_occ.columns = ['num_people']
hombres_occ = hombres_occ.reset_index()
hombres_occ

#ahora para las mujeres
mask = users.loc[:,'gender'] == 'F'
users_mujer = users.loc[mask,:]
mujer_occ = users_mujer.groupby('occupation')['occupation'].count().to_frame()
mujer_occ.columns = ['num_people']
mujer_occ = mujer_occ.reset_index()
mujer_occ

#fusionamos los tres data_frames anteriores y aÃ±adimos las columnas percentage
df1 = pd.merge(people_occ,hombres_occ,left_on='occupation', right_on='occupation',how='inner')
df_final = pd.merge(df1,mujer_occ,left_on='occupation', right_on='occupation',how='inner')
df_final.columns = ['occupation', 'total_people', 'males', 'females']
df_final
df_final.loc[:,'percentage_male'] = df_final.loc[:,'males'].div(df_final.loc[:,'total_people'])
df_final.loc[:,'percentage_female'] = df_final.loc[:,'females'].div(df_final.loc[:,'total_people'])
df_final

```
